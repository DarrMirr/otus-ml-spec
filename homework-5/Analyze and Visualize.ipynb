### Создайте одномерный массив размера 10, заполненный нулями и пятым элемент равным 1. Трансформируйте в двумерный массив.

### Создайте одномерный массив со значениями от 10 до 49 и разверните его (первый элемент становится последним). Найдите в нем все четные элементы.

### Создайте массив 4x3x2 со случайными значениями. Найти его минимум и максимум.

```python
a = np.random.randint(0, 50, size=(4,3,2))

max_element = np.max(a)
max_element_coord = np.where(a == max_element)
min_element = np.min(a)
min_element_coord = np.where(a == min_element)

print(a.shape)
print(a)
print(f'max_element={a[max_element_coord][0]}')
print(f'min_element={a[min_element_coord][0]}')
```

### Создайте два двумерных массива размерами 6x4 и 4x3 и произведите их матричное умножение. 

```python
a = np.random.randint(0, 3, size=(6,4))
b = np.random.randint(0, 3, size=(4,3))
c = a.dot(b)
print(f'a={a}')
print(f'b={b}')
print(f'c.shape={c.shape}')
print(f'c={c}')
```

### Создайте случайный двумерный массив 7x7, найти у него среднее и стандартное оклонение. Нормализуйте этот массив.

```python
a = np.random.rand(7, 7)
std = np.std(a)
avg = np.average(a)
norm_a = a / np.linalg.norm(a)
print(a)
print(f'standard deviation = {std}')
print(f'average = {avg}')
print(f'normilized a = {norm_a}')
```

# II. Pandas

### Импортируйте: pandas, matplotlib, seaborn

```python
import pandas as pd
import matplotlib
import seaborn as sns
```

### Загрузите датасет Tips из набора датасетов seaborn

```python
tips = sns.load_dataset("tips")
```

### Посмотрите на первые 5 строчек

```python
tips.head()
```

### Узнайте сколько всего строчек и колонок в данных

```python
tips.shape
```

### Проверьте есть ли пропуски в данных

```python
df_bool = tips.isna()
na_data = df_bool.where(df_bool == True)
print(na_data.dropna())
```

### Посмотрите на распределение числовых признаков

```python
tips.describe()
```

### Найдите максимальное значение 'total_bill'

```python
tips['total_bill'].max()
```

### Найдите количество курящих людей

```python
tips['smoker'].count()
```

### Узнайте какой средний 'total_bill' в зависимости от 'day'

```python
tips_group_by_day = tips.groupby('day', observed=False)
avg_total_bill_by_day = tips_group_by_day['total_bill'].mean()
print(pd.DataFrame(avg_total_bill_by_day))
```

### Отберите строчки с 'total_bill' больше медианы и узнайте какой средний 'tip' в зависимости от 'sex'

```python
total_bill_mean = tips['total_bill'].mean()
tips_above_mean = tips.where(tips['total_bill'] > total_bill_mean).dropna()
tam_group_by_sex = tips_above_mean.groupby('sex', observed=False)
avg_tam_group_by_sex = tam_group_by_sex['tip'].mean()
print(pd.DataFrame(avg_tam_group_by_sex))
```

### Преобразуйте признак 'smoker' в бинарный (0-No, 1-Yes)

```python
tips['smoker_binary'] = tips.smoker.map({'Yes': 1, 'No': 0})
print(tips[['smoker', 'smoker_binary']])
```

### Постройте два графика scatterplot, представляющих взаимосвязь между признаками 'total_bill' и 'tip' один для Male, другой для Female и раскрасьте точки в зависимоти от признака 'smoker'

```python
male_tips = tips[tips['sex']=='Male']
female_tips = tips[tips['sex']=='Female']
sns.scatterplot(data=male_tips, x='total_bill', y='tip', hue='smoker', style='smoker', palette='mako');
sns.scatterplot(data=female_tips, x='total_bill', y='tip', hue='smoker', style='smoker', palette='flare');
```
